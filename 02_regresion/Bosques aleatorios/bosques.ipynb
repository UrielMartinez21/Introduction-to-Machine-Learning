## Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Tratamiento de dataset

```python
# --> Importar dataset
resultado = pd.read_csv('Position_Salaries.csv')
resultado.head()

# --> Dividir dataset en variables independientes y dependientes
x = resultado.iloc[:, 1:2].values
y = resultado.iloc[:, 2].values
```

## Ajustar modelo a dataset

```python
from sklearn.ensemble import RandomForestRegressor
# --> Crear objeto de regresion
regression = RandomForestRegressor(n_estimators=300, random_state=0)

# --> Ajustar el modelo
regression.fit(x, y)
```

## Predicción

```python
y_pred = regression.predict([[6.5]])
y_pred
```

## Graficar resultados

```python
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))

plt.scatter(x, y, color='red')
plt.plot(x_grid, regression.predict(x_grid), color='blue')

plt.title('Modelo de Regresión')
plt.xlabel('Posición del empleado')
plt.ylabel('Sueldo (en $)')

plt.show()
```
