## Bibliotecas de inicio

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Visualizar dataset

```python
dataset = pd.read_csv("Position_Salaries.csv")
dataset.head()
```

## Separar en rasgos y clases

```python
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

## Ajustar dataset a regresión lineal

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

## Ajustar dataset a la regresion polinómica

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
```

## Nuevo modelo de regresión lineal

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

## Graficar modelo lineal

```python
# --> Graficar puntos
plt.scatter(x, y, color="red")
plt.plot(x, lin_reg.predict(x),color="blue")

# --> Personalizar gráfica
plt.title("Modelo lineal")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo($)")

# --> Mostrar gráfica
plt.show()
```

## Graficar modelo polinomial

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)

# --> Graficar puntos
plt.scatter(x, y, color="red")
plt.plot(x_grid, lin_reg2.predict(poly_reg.fit_transform(x_grid)), color="blue")

# --> Personalizar gráfica
plt.title("Modelo polinomial")
plt.xlabel("Posición de empleados")
plt.ylabel("Sueldo($)")

# --> Mostrar gráfica
plt.show()
```
