## Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Visualizar dataset

```python
resultado = pd.read_csv('Salary_Data.csv')
resultado.head()
```

## Separar en rasgos y clases

```python
x = resultado.iloc[:, :-1].values
y = resultado.iloc[:, 1].values
```

## Entranamiento y pruebas

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

## Modelo de regresión lineal

#### Crear modelo

```python
from sklearn.linear_model import LinearRegression

regresion = LinearRegression()
regresion.fit(x_train, y_train)
```

#### Predecir modelo

```python
y_predict = regresion.predict(x_test)
y_predict
```

## Visualizar los datos

#### Entranamiento

```python
plt.scatter(x_train, y_train, color='purple')                           # Puntos de entrenamiento
plt.plot(x_train, regresion.predict(x_train), color='blue')             # Recta de regresión

plt.title('Sueldo vs Años de experiencia (Conjunto de entrenamiento)')  # Título
plt.xlabel('Años de experiencia')                                       # Etiqueta eje X
plt.ylabel('Sueldo (en $)')                                             # Etiqueta eje Y

plt.show()                                                              # Mostrar gráfica
```

#### Prueba

```python
plt.scatter(x_test, y_test, color='purple')                             # Puntos de entrenamiento
plt.plot(x_train, regresion.predict(x_train), color='blue')             # Recta de regresión
plt.title('Sueldo vs Años de experiencia (Conjunto de prueba)')         # Título
plt.xlabel('Años de experiencia')                                       # Etiqueta eje X
plt.ylabel('Sueldo (en $)')                                             # Etiqueta eje Y
plt.show()                                                              # Mostrar gráfica
```
